```python
import pandas as pd

trdata = pd.read_csv('train.csv')
tedata = pd.read_csv('test.csv')

data = trdata + tedata
df = pd.DataFrame(data)
df.head()
```

```python
#drop id

df = df.drop(['PassengerId', 'Name'], axis = 1)
df.describe()

#there seems to be 177 missing 'Age' in the dataset
```

```python
df = df.dropna(subset=['Age'])
df.describe()
```

```python
import seaborn as sns

#lets look at whether male/female tend to survive
#there seems to be more female survivors than male, was the scene "save the women and children!" real?

sns.countplot(data=df, x="Sex", hue="Survived").set_title("1 = Survived")
```

```python
#lets find out if there are any patterns in "Age" and "Survivability"

sns.histplot(data=df, x="Age", hue='Survived', multiple='stack').set_title("1 = Survived")

#there are more survival for those between 10 to 40 y/o, however, there are also more individual within this age group on-board
#from the previous countplot, there seems to be more female than male survivor, lets plot female age plot to find out whether
#in general there are more females within 10 - 40 age than male which may lead to more female survivors
```

```python
gk = df.groupby('Sex')

female = gk.get_group('female')

sns.histplot(data=female, x='Age', multiple = 'stack')
```

```python
male = gk.get_group('male')

sns.histplot(data=male, x='Age', multiple = 'stack')
```

```python
#What about the rich? does those who pay more tend to also survive?

sns.catplot(data=df, x="Pclass", y="Survived", col="Sex",kind="bar")
#1st class seems to have a higher survival rate than 2nd and 3rd class passengers.
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr, vmax=0.5, annot=True, fmt=".2f", cmap='hot')
#there's a weak positive correlation between survival rate and fare
#there's a weak negatve correlation between survival rate and Pclass
#it makes sense Fare is positive correlation while Pclass is negative: First Class (Lower Pclass) = High Fare
```
